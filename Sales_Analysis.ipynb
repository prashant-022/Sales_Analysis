# Sales Analysis

```python
## Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Importing dataset:
df=pd.read_excel("superstore_sales.xlsx")
df.head()
```

```python
## Checking number of rows and columns:
df.shape
```

```python
## Checking if any null value:
df.isnull().sum()
```

From above, it can be seen that there is no null value in the sheet

```python
## Grouping the products and counting their numbers and arranging in descending order
broughts = df.groupby('product_name').agg({'quantity':'sum'}).sort_values('quantity',ascending=False)[:20]
broughts
```

```python
## Plotting the above data using barplot:
sns.barplot(x='product_name',y='quantity',data=broughts.reset_index())
plt.xticks(rotation='vertical')
```

```python
## Analysing sales of the products:
productSales = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['sales'])
productSales.sort_values(by=['sales'], inplace=True, ascending=False)
productSales.head(20)
```

```python
## Analysing profit data:
profit_product=pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['profit'])
profit_product.sort_values(by=['profit'],inplace=True, ascending=False)
profit_product.head(20)
```

```python
## Plotting profit vs product
sns.barplot(x='product_name',y='profit', data=profit_product.head(20).reset_index())
plt.xticks(rotation='vertical')
```

```python
df.groupby(['category','sales']).sum(numeric_only=True)
```

```python
## Analysing Profit on category basis:
profit_category = pd.DataFrame(df.groupby('category').sum(numeric_only=True)['profit'])
profit_category.sort_values(by=['profit'],inplace=True, ascending=False)
profit_category.head()
```

```python
## Plotting profit vs category plot:
sns.barplot(x='category', y='profit', data=profit_category.reset_index())
```

```python
## Plotting sales v/s category plot:
plt.figure(figsize=(10,5))
plt.bar(df['category'],df['sales'],color='b')
plt.xticks(size=12)
plt.show()
```

```python
## Analysing sub-category elements in their respective Category:
df.groupby(['category','sub_category']).sum(numeric_only=True)
```

```python
## Modifying the above data in more useful way:
df.groupby(['category','sub_category'],as_index=False)['quantity'].count()
```

```python
# Analyzing profit by segment data:
segment_profit = pd.DataFrame(df.groupby('segment').sum(numeric_only=True)['profit'])
segment_profit.sort_values(by=['profit'], inplace = True, ascending=False)
segment_profit
```

```python
## Plotting the segment v/s profit data:
sns.barplot(x='segment',y='profit',data=segment_profit.reset_index())
plt.xticks(rotation='vertical')
```

```python
## Analysis of Ship Mode:
ship_mode = pd.DataFrame(df.groupby('ship_mode').sum(numeric_only=True)['quantity'])
ship_mode.sort_values(by=['quantity'], inplace=True, ascending=False)
ship_mode
```

```python
## Plotting above ship_mode v/s quantity data:
sns.barplot(x='ship_mode', y='quantity',data=ship_mode.reset_index())
plt.xticks(rotation='horizontal')
```

```python
## Analysing the ship mode data with category:
category_hist = sns.FacetGrid(df,col='ship_mode',palette='rainbow')
category_hist.map(plt.hist,'category')
```

```python
## Analyzing counts of market:
sns.set_style('whitegrid')
sns.countplot(df,x='market',palette='rainbow')
```

```python
## Analysing sales by country:
country_sales = pd.DataFrame(df.groupby('country').sum(numeric_only=True)['sales'])
country_sales.sort_values(by=['sales'], inplace=True,ascending=False)
country_sales.head(20)
```

```python
## Analysis of shipping cost by country:
df.groupby('country').agg({'shipping_cost':'mean'}).sort_values(by=['shipping_cost'],ascending=False).head(20)
```

```python
data = df.sort_values(by='profit', ascending=False).head(10)[['order_id','customer_name','profit']]
data
```

```python
## Extracting year and month using datetime function and concating in the dataframe:
month = df['order_date'].astype(str)
month = month.apply(lambda x:datetime.strptime(x, '%Y-%m-%d'))
df['month'] = month.map(lambda x: x.month)
df
```

```python
date = df.groupby(['year','month']).sum(numeric_only=True)
date
```

```python
## Plotting the above year v/s sales data:
sns.barplot(x='year',y='sales',data=date.reset_index())
plt.xticks(rotation='vertical')
```
